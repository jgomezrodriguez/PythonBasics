**Plotting**

To plot mutltiple graphs in the same figure you can use the function hold, in case is desired to plot in another graph, you may do the use of plt.figure() before plotting the second plot, in this way, both will be plotted. 

```python
h = np.zeros(4) #A numpy array of zeros with lenght 4.
h[0] = 1.60; h[1] = 1.85; h[2] = 1.75; h[3] = 1.80 
H = np.zeros(4)
H[0] = 0.50; H[1] = 0.70; H[2] = 1.90; H[3] = 1.75
family_member_no = np.zeros(4)
family_member_no[0] = 0; family_member_no[1] = 1 #;to do in same line assignments
family_member_no[2] = 2; family_member_no[3] = 3
plt.plot(family_member_no, h, family_member_no, H)#it is alternating x and y values
plt.xlabel('Family member number')
plt.ylabel('Height (m)')
plt.show()
```

```python
#Using the hold on and hold off command
h = np.zeros(4) #A numpy array of zeros with lenght 4.
h[0] = 1.60; h[1] = 1.85; h[2] = 1.75; h[3] = 1.80 
H = np.zeros(4)
H[0] = 0.50; H[1] = 0.70; H[2] = 1.90; H[3] = 1.75
L = np.array([1.05,2.0,1.35,1.46])
family_member_no = np.zeros(4)
family_member_no[0] = 0; family_member_no[1] = 1 #;to do in same line assignments
family_member_no[2] = 2; family_member_no[3] = 3
plt.plot(family_member_no,h,'go',label='label1')#Assign the label here for legend
plt.plot(family_member_no,H,'ro',label='label2')#Using these ro,bo,go for markers
plt.plot(family_member_no,L,'bo',label='label3')
plt.xlabel('Family member number')
plt.legend()
plt.ylabel('Height (m)')
plt.show()
```

In case we want to bound the axis in the x and y directions and we want to create all the plants in different figures, we can ask python to do the plt.figure() after the script to plot each.

```python
#Using the hold on and hold off command
h = np.zeros(4) #A numpy array of zeros with lenght 4.
h[0] = 1.60; h[1] = 1.85; h[2] = 1.75; h[3] = 1.80 
H = np.zeros(4)
H[0] = 0.50; H[1] = 0.70; H[2] = 1.90; H[3] = 1.75
L = np.array([1.05,2.0,1.35,1.46])
family_member_no = np.zeros(4)
family_member_no[0] = 0; family_member_no[1] = 1 #;to do in same line assignments
family_member_no[2] = 2; family_member_no[3] = 3
plt.plot(family_member_no,h,'go',label='label1')
plt.xlabel('Family member number')
plt.legend()
plt.ylabel('Height (m)')#Assign the label here for legend
plt.figure()
plt.plot(family_member_no,H,'ro',label='label2')
plt.xlabel('Family member number')
plt.legend()
plt.ylabel('Height (m)')
plt.figure()#Using these ro,bo,go for markers
plt.plot(family_member_no,L,'bo',label='label3')
plt.xlabel('Family member number')
plt.legend()
plt.ylabel('Height (m)')
plt.axis([1,2,1,2]) #plt.axis([xmin,xmax,ymin,ymax]) #LImitation on the third figure axis
plt.figure()
```

**Creation of matrix and vectors** <br>
The use of numpay module ensures the creation of vectors and matrices. <br>
Note that the matrix are defined as (rows,columns) and the vectors are row vectors. 

```python
x = np.zeros(2)
x = np.mat(x)#It will treat the object as a linear algebra object (vector)
x = np.transpose(x)#It will traspose the row vector into a column vector
x[0] = 3; x[1] = 2
A = np.zeros((2,2))#Array of 2x2
A = np.mat(A)
A[0,0] = 1; A[0,1] = 0 #The utility of the ; in case of multiple variables assignments
A[1,0] = 0; A[1,1] = 1
#Now, we are simply multiplying the matrix by the vector already assigned to x
y = A*x
print(y)
```

**Symbolic representation of the variables**\
The module sympy allows us to write without any computation the variables x,y,z, etc.

```python
from sympy import *
x,y = symbols('x y')
z = x*y
print(z)
```

```python
from sympy import * #It contains multiple mathematical operations
x = Symbol('x')#It creates the symbolic variable x and y of the mathematical domain
y = Symbol('y')
print (2*x + 3*x - y)# Algebraic computation
print (diff(x**2, x))# Differentiates x**2 wrt. x
print (integrate(cos(x), x))# Integrates cos(x) wrt. x
print (simplify((x**2 + x**3)/x**2))# Simplifies expression
print (limit(sin(x)/x, x, 0))# Finds limit of sin(x)/x as x->0
print (solve(5*x - 15, x))# Solves 5*x = 15
```
